```python
import numpy as np
import matplotlib.pyplot as plt
```

## Bin Step

```python
def binstep(x):
    return np.heaviside(x,1)
x = np.linspace(-10,10)
plt.plot(x,binstep(x))
plt.axis('tight')
plt.title('BinStep')
plt.show()
```

## Linear

```python
def linear(x):
    return x
x = np.linspace(-10,10)
plt.plot(x,linear(x))
plt.axis('tight')
plt.title('Linear')
plt.show()
```

## Sigmoid

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('sigmoid')
plt.show()
```

## Bipolar Sigmoid

```python
def bipsig(x):
    return -1+2/(1+np.exp(-x))
x = np.linspace(-10,10)
plt.plot(x,bipsig(x))
plt.axis('tight')
plt.title('Bipolar Sigmoid')
plt.show()
```

## Tanh

```python
def tanh(x):
    return np.tanh(x)
x = np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title('Tanh')
plt.show()
```

## Relu

```python
def relu(x):
    x1 = []
    for i in x:
        if i < 0:
            x1.append(0)
        else:
            x1.append(i)
    return x1

x = np.linspace(-10, 10)
y = relu(x)  # Compute ReLU values for all elements of x

plt.plot(x, y)
plt.axis('tight')
plt.title('ReLU Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

## Leaky ReLU

```python
def LeakyRelu(x):
 alpha = 0.1
 return np.maximum(alpha*x, x)
x = np.linspace(-10, 10)
plt.plot(x, LeakyRelu(x))
plt.axis('tight')
plt.title('Activation Function :Leaky Relu')
plt.show()
```

## Softmax

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x),axis=0)
x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.show()
```
