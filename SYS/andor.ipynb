```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import array , random, dot
from pylab import ylim,plot
from random import choice
```

## OR

```python
step_func = lambda x:0 if x<0 else 1
data = [
    (array([0,0,1]),0),
    (array([0,1,1]), 1), 
    (array([1,0,1]), 1), 
    (array([1,1,1]), 1),
]
weights = np.random.rand(3)
error = []
lr = 0.1
n = 100
```

```python
for j in range(n):
    x,target = choice(data)
    result = dot(weights,x)
    err = target - step_func(result)
    error.append(err)
    weights += lr *err *x
```

```python
for x, _ in data:
    result = dot(x,weights)
    print(f"{x}: {result} -> {step_func(result)}")
ylim([-1,1])
plt.plot(error)
plt.show()
```
