```python
import string

```

```python
def enc(word):
    wl = []
    for i in word:
        if i in string.ascii_lowercase:
            alph = string.ascii_lowercase.index(i)
            w = (alph + 3) % 26
            strn = string.ascii_lowercase[w]
            wl.append(strn)
        else:
            wl.append(i)  # Keep non-alphabetic characters unchanged
    return "".join(wl)

```

```python
def dec(word):
    wl = []
    for i in word:
        if i in string.ascii_lowercase:
            alph = string.ascii_lowercase.index(i)
            w = (alph - 3) % 26
            strn = string.ascii_lowercase[w]
            wl.append(strn)
        else:
            wl.append(i)  # Keep non-alphabetic characters unchanged
    return "".join(wl)

```

```python
def create_matrix(text, columns):
    rows = (len(text) + columns - 1) // columns
    matrix = [['' for _ in range(columns)] for _ in range(rows)]
    index = 0
    for row in range(rows):
        for col in range(columns):
            if index < len(text):
                matrix[row][col] = text[index]
                index += 1
    return matrix

```

```python
def print_matrix(matrix):
    for row in matrix:
        print(''.join(row))

```

```python
def permute_columns(matrix, key):
    columns = len(matrix[0])
    rows = len(matrix)
    permuted_matrix = [['' for _ in range(columns)] for _ in range(rows)]

    # Create a list of column indices and sort them based on the key
    column_indices = list(range(columns))
    sorted_indices = sorted(column_indices, key=lambda x: key.index(x))

    for row in range(rows):
        for col, sorted_col in enumerate(sorted_indices):
            permuted_matrix[row][col] = matrix[row][sorted_col]

    return permuted_matrix

```

```python
def main():
    # Step 2
    text = input("Enter the string to be encrypted: ").lower()

    # Step 3
    k = int(input("Input the k for Caesar cipher (this will be ignored, fixed shift of 3 is used): "))

    # Step 4
    encrypted_message = enc(text)
    print(f"Encrypted message: {encrypted_message}")

    # Step 5
    columns = int(input("Input the number of columns: "))
    rows = (len(encrypted_message) + columns - 1) // columns

    # Step 6
    matrix = create_matrix(encrypted_message, columns)
    print("Matrix form:")
    print_matrix(matrix)

    # Step 7
    key_str = input("Enter the encryption key (comma-separated column order): ")
    key = list(map(int, key_str.split(',')))

    # Step 8
    permuted_matrix = permute_columns(matrix, key)
    print("Permuted matrix:")
    print_matrix(permuted_matrix)

    # Decryption process (for verification)
    flat_permuted_message = ''.join([''.join(row) for row in permuted_matrix])
    decrypted_message = dec(flat_permuted_message)
    print(f"Decrypted message (for verification): {decrypted_message}")

if __name__ == "__main__":
    main()
```
